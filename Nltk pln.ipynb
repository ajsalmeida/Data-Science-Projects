```python
import nltk #biblioteca de processamento de linguagem natural
import requests #biblioteca de requisições
from bs4 import BeautifulSoup #webscrapping
import re #expressões regulares
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('stopwords')
nltk.download('punkt')
```

```python
page_uol = requests.get('https://uol.com.br')
soup_UOL = BeautifulSoup(page_uol.text, 'html.parser')
```

```python
manchetes_UOL = soup_UOL.find_all('h3') 
texto = ''
for i in manchetes_UOL:
    texto+=' '+i.get_text().strip()
```

```python
texto
```

```python
texto_min = texto.lower()
```

```python
palavras = re.findall(r'[a-àzéíúóáêâãõôç]+',texto_min)
```

```python
texto_unido = " ".join(palavras)
```

```python
texto_unido
```

```python
texto_tokens = word_tokenize(texto_unido)
```

```python
texto_tokens
```
